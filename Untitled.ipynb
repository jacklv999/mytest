```python
## create time stamp
t_n = time.localtime(time.time())
time_now_1 = str('0'+str(t_n[1])) if t_n[1]<10 else str(t_n[1])
time_now_2 = str('0'+str(t_n[2])) if t_n[2]<10 else str(t_n[2])
time_now = time_now_1+time_now_2
```
